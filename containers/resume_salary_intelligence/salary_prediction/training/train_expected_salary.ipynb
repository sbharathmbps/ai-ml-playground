# Outlier Treatment

```python
# plt.figure(figsize=(15,15))
# sns.boxplot(data)
# plt.xticks(rotation=90)
```

```python
for i in ["Total_Experience_in_field_applied","Certifications","International_degree_any"]:
    Q1=data[i].quantile(0.25)
    Q3=data[i].quantile(0.75)
    IQR=Q3-Q1
    data[i]=np.where(data[i]<Q1 - 1.5 * IQR,Q1,data[i])
    data[i]=np.where(data[i]>Q3 + 1.5 * IQR,Q3,data[i])
```

## Linear Regression:

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
regression_model = LinearRegression()
regression_model.fit(X_train5, y_train)
```

```python
print('The coefficient of determination R^2 of the prediction on Train set',regression_model.score(X_train5, y_train))
print('The coefficient of determination R^2 of the prediction on Test set',regression_model.score(X_test, y_test))
```

```python
print("The Root Mean Square Error (RMSE) of the model is for testing set is",np.sqrt(mean_squared_error(y_test,regression_model.predict(X_test))))
```
